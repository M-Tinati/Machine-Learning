```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#import data and check
data = pd.read_csv("../data/SalaryData.csv")
data.head()
```

```python
data.describe()
```

```python
#train & test
Data = pd.DataFrame(data)
X = Data[["YearsExperience"]]
y = Data["Salary"]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#model
model = LinearRegression()
model
```

```python
model.fit(x_train , y_train)
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print("Slope (coefficient): ", model.coef_[0])
print("Intercept: ", model.intercept_)

print("Predictions: ", y_predict)

print("R² score: ", r2_score(y_test, y_predict))
print("Mean Squared Error (MSE): ", mean_squared_error(y_test, y_predict))
```

```python
def chart():
    plt.scatter(x_test, y_test, color='royalblue', label='Actual Values')
    plt.plot(x_test, y_predict, color='tomato', label='Model Predictions')
    plt.xlabel("Years of Experience")
    plt.ylabel("Salary")
    plt.title("Linear Regression: Salary vs Experience")
    plt.legend()
    plt.grid(True)
    plt.show()
chart()
```
